```python
import pandas as pd
import plotly.express as px
```

```python
mdf = pd.read_csv("microbit.csv")
```

```python
mdf["strength"] = (mdf["x"]**2 + mdf["y"]**2 + mdf["z"]**2)**0.5
```

```python
fig = px.line(mdf, x="Time", y=["x", "y", "z", "strength"], title='microbit accelerometer plot', markers=True)
fig.show()
```

```python
mdf[(mdf["strength"]>2000) & (mdf["Time"]<45)].count()
```

```python

sdf["a1magChange"] = sdf["a1mag"].diff()
sdf["a2magChange"] = sdf["a2mag"].diff()
sdf["timeChange"] = sdf["time"].diff()
```

```python
sdf["a1magSpeed"] = sdf["a1magChange"] / sdf["timeChange"]
sdf["a2magSpeed"] = sdf["a2magChange"] / sdf["timeChange"]
```

```python
sdf["magS"] = sdf["a1mag"] + sdf["a2mag"]
sdf["magD"] = sdf["mags"]/6400
sdf["magDC"] = sdf["magD"].cumsum()
```

```python
sdf["magComplement"] = sdf["a1mag"]*sdf["a2mag"]/100000
sdf["magCompCum"] = sdf["magComplement"].cumsum()
```

```python
# fig = px.line(sdf, x="time", y=["a1", "a2", "a3", "a4", "a5", "a6"], title='play impossible accelerometer plot')
# fig = px.line(sdf, x="time", y=["a1mag", "a2mag", "mags", "magD"], title='play impossible accelerometer plot', markers=True)
# fig = px.line(sdf, x="time", y=["magCompCum", "magDC", "magS"], title='play impossible accelerometer plot', markers=True)
fig = px.line(sdf, x="time", y=["magS"], title='play impossible accelerometer plot', markers=True)
fig.show()
```

```python
sdf["a1shake"] = 0
sdf.loc[sdf['a1mag'] > 75, 'a1shake'] = sdf.loc[sdf['a1mag'] > 75, 'a1mag'] - 87
```

```python
sdf["a1sks"] = sdf["a1shake"].cumsum()
```

```python
fig = px.line(sdf, x="time", y=["a1sks"], title='play impossible accelerometer plot', markers=True)
fig.show()
```
